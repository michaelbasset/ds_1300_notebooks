```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('folktales.csv')
data.head()
```

```python
data = data.fillna('No Data')
```

```python
ax = data.plot.scatter(x='Year Written', y="Story Type", color='Red',label="Story Type", marker='^',figsize=(10,60))
data.plot.scatter(x='Year Written', y='Author', label="Author", ax=ax, xlabel="Year Written", ylabel="Story Info", title='Folktales');
```

My scatter plot shows not only the date a certain story was written, by also the author and genre for said book. Along the x-axis is the year written, and along the y-axis is the author and the genre type.

I chose this data set because I wanted to work with dates, so I have a set start start, ex for this one the start date was 1812. I wanted to have this number because I then knew where I needed to start. Also when we worked on the folktales in class I enjoyed it, so I chose to expand on it.
